```python
x_data=np.linspace(0,1,10)
y_data=np.linspace(1,2,10)
x_ext=6
y_ext=6
```

```python
fig = plt.figure(figsize=(x_ext, y_ext))
ax = fig.add_subplot(111)
extent = [np.min(x_data), np.max(x_data), np.min(y_data), np.max(y_data)]
```

```python
fig.show()
aspect_ratio = abs((x_data[-1] - x_data[0]) / (y_data[-1] - y_data[0]) )
```

```python
cp = ax.imshow(z_data, cmap='hot', 
                 interpolation='nearest',extent=extent, aspect=aspect_ratio)
```

```python
cb = fig.colorbar(cp, fraction=0.046/2, pad=0.04)
```

```python
fig.canvas.draw()
```

```python
fig.tight_layout()
```

```python
z_data *=2
```

```python
cp.set_data(z_data)
```

```python
cNorm      = matplotlib.colors.Normalize(vmin=np.min(z_data),vmax=np.max(z_data))
scalarMap  = matplotlib.cm.ScalarMappable(norm=cNorm, cmap=plt.get_cmap('hot') )
```

```python
cb.update_normal(scalarMap)
```

```python
cp.set_norm(cNorm)
```
